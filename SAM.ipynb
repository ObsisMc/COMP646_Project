```python


# Load the background image and the segment to blend
background = cv2.imread('background.jpg')
segment = cv2.imread('segment.png')
mask = np.zeros(background.shape[:2], dtype=np.uint8)
mask[100:400, 100:400] = 255  # Assuming the segment is placed at these coordinates

# Coordinates where to place the top-left corner of the segment on the background
x, y = 100, 100

# Perform Poisson blending
blended = cv2.seamlessClone(segment, background, mask, (x + int(segment.shape[1]/2), y + int(segment.shape[0]/2)), cv2.NORMAL_CLONE)

cv2.imshow('Blended Image', blended)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
